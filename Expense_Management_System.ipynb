```python
# expense_system.py

expenses = []  # List to store daily expense records
category_totals = {}  # Dictionary for category-wise totals

def add_expense():
    try:
        amount = float(input("Enter expense amount: "))
        if amount <= 0:
            raise ValueError("Amount must be positive.")
        category = input("Enter category (e.g., Food, Travel, Bills): ").strip()
        description = input("Enter description: ").strip()

        # Add to expenses list
        expense = {"amount": amount, "category": category, "description": description}
        expenses.append(expense)

        # Update category totals
        category_totals[category] = category_totals.get(category, 0) + amount

        print("Expense added successfully!")
    except ValueError as e:
        print(f"Invalid input: {e}")

def view_summary():
    print("\n--- Expense Summary ---")
    for category, total in category_totals.items():
        print(f"{category}: ₹{total:.2f}")

def export_data():
    try:
        with open("expenses.txt", "a") as file:
            for exp in expenses:
                file.write(f"{exp['amount']},{exp['category']},{exp['description']}\n")

        with open("summary.txt", "w") as file:
            for category, total in category_totals.items():
                file.write(f"{category}: ₹{total:.2f}\n")

        print("Data exported successfully!")
    except IOError as e:
        print(f"File error: {e}")

# Main Menu
while True:
    print("\nExpense Management System")
    print("1. Add Expense")
    print("2. View Summary")
    print("3. Export Data")
    print("4. Exit")

    choice = input("Enter your choice: ")
    if choice == "1":
        add_expense()
    elif choice == "2":
        view_summary()
    elif choice == "3":
        export_data()
    elif choice == "4":
        print("Exiting... Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
```
